```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Group by 'Month' and sum up the Revenue column
monthly_revenue = data.groupby('Month')['Revenue'].sum().reset_index()
monthly_revenue.columns = ['Month', 'Total_Revenue']  # Rename for clarity
```

```python
# Define the correct month order
month_order = ["January", "February", "March", "April", "May", "June",
               "July", "August", "September", "October", "November", "December"]

# Convert 'Month' to a categorical type with the specified order
monthly_revenue['Month'] = pd.Categorical(monthly_revenue['Month'], categories=month_order, ordered=True)
```

```python
# Plot total revenue per month
plt.figure(figsize=(10, 6))
sns.barplot(data=monthly_revenue, x='Month', y='Total_Revenue', color='skyblue')

# Add labels and title
plt.title('Monthly Revenue')
plt.xlabel('Month')
plt.ylabel('Total Revenue')

# Display the plot
plt.show()

#commentary: there are a lot more purchases in the holiday months...
```

```python
# Group by 'Month' and 'Revenue', then count occurrences
monthly_revenue_counts = data.groupby(['Month', 'Revenue']).size().unstack(fill_value=0).reset_index()
monthly_revenue_counts.columns = ['Month', 'No_Purchases', 'Purchases']  # Rename for clarity
```

```python
# Define the correct month order
month_order = ["January", "February", "March", "April", "May", "June",
               "July", "August", "September", "October", "November", "December"]

# Convert 'Month' to a categorical type with the specified order
monthly_revenue_counts['Month'] = pd.Categorical(monthly_revenue_counts['Month'], categories=month_order, ordered=True)
```

```python
# Melt the DataFrame for easier plotting with Seaborn
monthly_revenue_melted = monthly_revenue_counts.melt(id_vars='Month', value_vars=['No_Purchases', 'Purchases'],
                                                     var_name='Revenue_Type', value_name='Count')

# Plotting side-by-side bars for each month
plt.figure(figsize=(12, 6))
sns.barplot(data=monthly_revenue_melted, x='Month', y='Count', hue='Revenue_Type', palette={'No_Purchases': 'lightcoral', 'Purchases': 'skyblue'})

# Add labels and title
plt.title('Monthly Purchases vs. No Purchases')
plt.xlabel('Month')
plt.ylabel('Count')
plt.legend(title='Revenue Type', labels=['No Purchases', 'Purchases'])

# Display the plot
plt.show()

#commentary: probably will not include this plot (its not that helpful to see which months had a higher rate of no purchases being made)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Use the specified conversion rates
visitor_data = pd.DataFrame({
    'VisitorType': ['New Visitor', 'Other', 'Returning Visitor'],
    'ConversionRate': [24.91, 18.82, 13.93]
})

# Define custom colors similar to your R plot
colors = ["#c6e2c6", "#98fb98", "#2e8b57"]

# Create the bar plot
plt.figure(figsize=(8, 6))
sns.barplot(data=visitor_data, x='VisitorType', y='ConversionRate', palette=colors)

# Add title and labels
plt.title('Conversion Rates by Visitor Type')
plt.xlabel('Visitor Type')
plt.ylabel('Conversion Rate (%)')

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right')

# Apply a minimal theme
sns.set_theme(style="whitegrid")

# Show the plot
plt.show()


#commentary: This plot shows the conversion rates by visitor type, indicating the percentage of visitors in each category
#(New Visitor, Other, Returning Visitor) who completed a desired action. The results suggest that New Visitors have the highest
#conversion rate at 24.91%, followed by Other visitors at 18.82%, while Returning Visitors have the lowest conversion rate at 13.93%.
#This insight is significant as it highlights that new visitors are more likely to convert than returning visitors, which could inform targeted
#marketing or engagement strategies aimed at increasing conversions among different visitor segments.
```

```python
# Group by Revenue and calculate average visits and durations
behavior_analysis = data.groupby('Revenue').agg({
    'Administrative': 'mean',
    'Informational': 'mean',
    'ProductRelated': 'mean',
    'Administrative_Duration': 'mean',
    'Informational_Duration': 'mean',
    'ProductRelated_Duration': 'mean'
}).reset_index()

# Rename columns for clarity
behavior_analysis.columns = [
    'Revenue', 'Avg_Admin_Visits', 'Avg_Info_Visits', 'Avg_Product_Visits',
    'Avg_Admin_Duration', 'Avg_Info_Duration', 'Avg_Product_Duration'
]
```

```python
# Melt the visits data
data_visits = behavior_analysis.melt(
    id_vars='Revenue',
    value_vars=['Avg_Admin_Visits', 'Avg_Info_Visits', 'Avg_Product_Visits'],
    var_name='Metric',
    value_name='Value'
)

# Map the metric names to types
data_visits['Type'] = data_visits['Metric'].replace({
    'Avg_Admin_Visits': 'Admin',
    'Avg_Info_Visits': 'Info',
    'Avg_Product_Visits': 'Product'
})
```

```python
# Melt the duration data
data_duration = behavior_analysis.melt(
    id_vars='Revenue',
    value_vars=['Avg_Admin_Duration', 'Avg_Info_Duration', 'Avg_Product_Duration'],
    var_name='Metric',
    value_name='Value'
)

# Map the metric names to types
data_duration['Type'] = data_duration['Metric'].replace({
    'Avg_Admin_Duration': 'Admin',
    'Avg_Info_Duration': 'Info',
    'Avg_Product_Duration': 'Product'
})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting average visits
plt.figure(figsize=(10, 6))
sns.barplot(data=data_visits, x='Type', y='Value', hue='Revenue', palette=['salmon', 'turquoise'])
plt.title('Average Page Visits by Page Type and Conversion')
plt.xlabel('Page Type')
plt.ylabel('Average Visits')
plt.legend(title='Conversion (0 = No, 1 = Yes)')
plt.show()

#commentary: This plot shows that Product-related pages receive significantly more visits on average than both Administrative and Informational pages, particularly
#when there is a conversion (Revenue = 1). It indicates that users who end up converting spend more time engaging with product-related content compared to other page types,
#suggesting these pages play a key role in driving conversions.
```

```python
# Plotting average duration
plt.figure(figsize=(10, 6))
sns.barplot(data=data_duration, x='Type', y='Value', hue='Revenue', palette=['salmon', 'turquoise'])
plt.title('Average Time Spent by Page Type and Conversion')
plt.xlabel('Page Type')
plt.ylabel('Average Duration (seconds)')
plt.legend(title='Conversion (0 = No, 1 = Yes)')
plt.show()

#commentary: This plot shows that users spend the most time on Product-related pages, especially when they convert (Revenue = 1), with average durations far exceeding time spent
#on Administrative or Informational pages. This suggests that high engagement with product pages is strongly associated with conversion, highlighting the importance of these pages
#in influencing purchasing behavior.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# One-hot encode categorical variables
data_encoded = pd.get_dummies(data, drop_first=True)

# Calculate correlation matrix
correlation_matrix = data_encoded.corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap (Including Encoded Categorical Variables)")

plt.savefig("correlation_heatmap.png")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set up the matplotlib figure with a larger size
plt.figure(figsize=(14, 12))

# Define a threshold for annotating the correlation values
threshold = 0.2

# Plot the heatmap with improved readability
sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5,
    linecolor='white',
    annot_kws={"size": 8},  # Font size for annotations
    mask=abs(correlation_matrix) < threshold,  # Mask to hide values below threshold
    cbar_kws={"shrink": .8}  # Shrink color bar size
)

# Customize the title and label sizes
plt.title("Improved Correlation Heatmap", fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate and set font size for x labels
plt.yticks(fontsize=10)  # Set font size for y labels

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping

plt.savefig("improved_correlation_heatmap.png")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set up the matplotlib figure with a larger size for readability
plt.figure(figsize=(14, 12))

# Plot the heatmap with hierarchical clustering enabled
sns.clustermap(
    correlation_matrix,
    annot=True,
    fmt=".1f",
    cmap="coolwarm",
    linewidths=0.5,
    linecolor='gray',
    annot_kws={"size": 8},  # Smaller font size for readability
    cbar_kws={"shrink": 0.8},  # Shrink color bar
    dendrogram_ratio=(.1, .2),  # Adjust dendrogram size
    figsize=(14, 12)  # Figure size
)

# Show the clustered heatmap
plt.suptitle("Clustered Correlation Heatmap", y=1.05, fontsize=16)

plt.savefig("Clustered Correlation Heatmap.png")

plt.show()
```
