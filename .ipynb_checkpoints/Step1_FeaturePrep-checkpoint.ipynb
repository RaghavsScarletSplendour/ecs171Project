```python
import pandas as pd

# Load the dataset
data_path = 'ECS171 Project Dataset.csv'  # Update with the correct path if needed
data = pd.read_csv(data_path)

# Drop specified columns
columns_to_drop = [
    "Administrative", "Administrative_Duration",
    "Informational", "Informational_Duration",
    "PageValues", "SpecialDay"
]
data_cleaned = data.drop(columns=columns_to_drop)

# Map months to quarters
month_to_quarter = {
    "Feb": "Q1", "Mar": "Q1", "Apr": "Q1",
    "May": "Q2", "Jun": "Q2", "Jul": "Q2",
    "Aug": "Q3", "Sep": "Q3", "Oct": "Q3",
    "Nov": "Q4", "Dec": "Q4"
}
data_cleaned["Quarter"] = data_cleaned["Month"].map(month_to_quarter)

# Save or use the cleaned dataset
data_cleaned.head()  # Display the first few rows
```

```python
# Save the cleaned dataset to a new file
output_path = 'Cleaned_ECS171_Dataset.csv'
data_cleaned.to_csv(output_path, index=False)

print(f"Cleaned dataset saved to {output_path}")
```
